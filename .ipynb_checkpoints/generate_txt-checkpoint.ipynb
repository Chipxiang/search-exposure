```python
import csv
import re
import random
import math
import numpy as np
```

```python
# Generate .txt file to run pretrained DSSM model
query_path = "msmarco-test2019-queries.tsv"
rel_path = "2019qrels-pass.txt"
passage_path = "collection.tsv"
folder_path = "/home/jianx/data/"
out_file_path = "./pairs.txt"
```

```python
with open(folder_path + query_path, mode = 'r') as infile:
    rows = (line.split('\t') for line in infile)
    re.sub("[^a-zA-Z]+", "", "ABC12abc345def")
    q_dict = {row[0]:re.sub("[^a-zA-Z]+", " ", row[1:][0].replace('\n', '')) for row in rows}
```

```python
def load_dict(path, folder = folder_path):
    with open(folder + path, mode = 'r') as infile:
        rows = (line.split('\t') for line in infile)
        out_dict = {row[0]:re.sub("[^a-zA-Z]+", " ", row[1:][0].replace('\n', '')) for row in rows}
    return out_dict
```

```python
q_dict = load_dict(query_path)
```

```python
import pickle
def obj_reader(path):
    with open(path, 'rb') as handle:
        return pickle.loads(handle.read())
RATING_DICT_PATH = "/home/jianx/data/rel_scores.dict"
rel_dict = obj_reader(RATING_DICT_PATH)
```

```python
TOP_DICT_PATH = "/home/jianx/data/initial_ranking.dict"
top_dict = obj_reader(TOP_DICT_PATH)
```

```python
p_dict = load_dict(passage_path)
```

```python
qids = list(rel_dict.keys())
with open(out_file_path, 'w') as f:
    for i, qid in enumerate(qids):
        query = q_dict[str(qid)]
        candidates = top_dict[qid]
        for pid in candidates:
            passage = p_dict[str(pid)]
            f.write(query + "\t" + passage + "\n")
```

```python
def get_ndcg_precision_rr(true_dict, test_dict, rank):
    sorted_result = sorted(test_dict.items(), key=lambda x: (x[1], [-1,1][random.randrange(2)]), reverse=True)
    original_rank = rank
    rank = min(rank, len(sorted_result))
    cumulative_gain = 0
    num_positive = 0
    rr = float("NaN")
    for i in range(len(sorted_result)):
        pid = sorted_result[i][0]
        if pid in true_dict:
            rr = 1 / (i + 1)
            break
    for i in range(rank):
        pid = sorted_result[i][0]
        if pid in true_dict:
            num_positive += 1
    sorted_result = sorted(test_dict.items(), key=lambda x: x[1], reverse=True)
    for i in range(rank):
        pid = sorted_result[i][0]
        relevance = 0
        if pid in true_dict:
            relevance = true_dict[pid]
        discounted_gain = relevance / math.log2(2 + i)
        cumulative_gain += discounted_gain
    sorted_ideal = sorted(true_dict.items(), key=lambda x: x[1], reverse=True)
    ideal_gain = 0
    for i in range(rank):
        relevance = 0
        if i < len(sorted_ideal):
            relevance = sorted_ideal[i][1]
        discounted_gain = relevance / math.log2(2 + i)
        ideal_gain += discounted_gain
    ndcg = 0
    if ideal_gain != 0:
         ndcg = cumulative_gain / ideal_gain
    return ndcg, num_positive / original_rank, rr
```

```python
out_scores = []
with open("dssm_out.score.txt", mode='r') as infile:
    reader = csv.reader(infile)
    for rows in reader:
        out_scores.append(float(rows[0]))
```

```python
result_dict = {}
count = 0
for i, qid in enumerate(qids):
    candidates = top_dict[qid]
    q_results = {}
    for pid in candidates:
        q_results[pid] = out_scores[count]
        count += 1
    result_dict[qid] = q_results
```

```python
def evaluation(relevance, results):
    result_ndcg = []
    result_prec = []
    result_rr = []
    qids = list(relevance.keys())
    for qid in qids:
        ndcg, prec, rr = get_ndcg_precision_rr(relevance[qid], results[qid], rank = 10)
        result_ndcg.append(ndcg)
        result_prec.append(prec)
        result_rr.append(rr)
    avg_ndcg = np.mean(result_ndcg)
    avg_prec = np.mean(result_prec)
    avg_rr = np.mean(result_rr)
    return result_ndcg, result_prec, result_rr, avg_ndcg, avg_prec, avg_rr
```

```python
ndcgs, precs, rrs, avgndcg, avgprec, avgrr = evaluation(rel_dict, result_dict)
```
